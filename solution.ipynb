```python
# Get the data
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 

data = pd.read_csv('./files_for_lab/csv_files/marketing_customer_analysis.csv')
data.head()
```

```python
#Show DataFrame info.
data.info()
```

```python
#Describe DataFrame.
data.describe()
```

```python
#Show a plot of the total number of responses.
fig, ax = plt.subplots(figsize=(5,5), dpi=100)
ax.hist(data['Response'])
ax.set_title('Total Responses')
plt.show()
```

```python
#Show a plot of the response rate by the sales channel. And filter out the 'Responses' with 'No' value. 

fig, ax = plt.subplots(figsize=(5,5), dpi=100)
ax.hist(data.loc[data['Response'] == 'Yes', 'Sales Channel']) 
ax.set_title('Response Rate by Sales Channel') 
plt.show() 
```

```python
#Show a plot of the response rate by the total claim amount.
fig, ax = plt.subplots(figsize=(5,5), dpi=100) 
ax.hist(data.loc[data['Response'] == 'Yes', 'Total Claim Amount']) 
ax.set_title('Total Responses per Total Claim Amount')
ax.set_xlabel('Total Claim Amount')
ax.set_ylabel('Total Responses')
plt.show() 
```

```python
#Show a plot of the response rate by income.
fig, ax = plt.subplots(figsize=(5,5), dpi=100) 
ax.hist(data.loc[data['Response'] == 'Yes', 'Income']) 
ax.set_title('Total Responses per Income') 
ax.set_xlabel('Income') 
ax.set_ylabel('Total Responses') 
plt.show() 
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(12,6)) 
fig.suptitle('Total Responses per Income') #This is the title for the whole figure

ax1.hist(data.loc[data['Response'] == 'Yes', 'Income']) 
ax1.set_xlabel('Income') 
ax1.set_ylabel('Response=Yes') 

ax2.hist(data.loc[data['Response'] == 'No', 'Income']) 
ax2.set_xlabel('Income') 
ax2.set_ylabel('Response=No') 

plt.show()
```
